```python
images = []
names = []
yop = []
catagory = []

for j in range(1,5):
  webpage = requests.get(f"https://www.amazon.in/s?k=poetry+books&page={j}",headers=headers).text
  soup = BeautifulSoup(webpage,"lxml")
  infoDiv = soup.find_all("div",class_="puis-card-container")
  for i in infoDiv:
    img_tag = i.find("img", class_="s-image")
    name  = i.find("span",class_="a-text-normal").text.strip()
    year = int(i.find("span",class_="a-size-base a-color-secondary a-text-normal").text.strip().split(" ")[-1]) if i.find("span",class_="a-size-base a-color-secondary a-text-normal") else 0
    if img_tag:
      src_value = img_tag.get("src")
      images.append(src_value)
      names.append(name)
      yop.append(year)
      catagory.append(0)
  else:
      print("No img tag with class 's-image' found in this div.")


```
